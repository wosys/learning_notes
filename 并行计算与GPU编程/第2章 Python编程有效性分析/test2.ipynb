```python
# 
import warnings
warnings.filterwarnings('ignore')

import time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def iterator(c,r,max_iter):
    '''
    
    crmax_iter
    
    '''
    z=c #
    for iter in range(0,max_iter,1): 
        if abs(z)>r:break
        z=z**2+c 
    return iter 

start_time = time.time()

#mandelbrot 
X=np.linspace(-1.75,1.05,5000) #5000
Y=np.linspace(-1.25,1.25,5000) #5000
real,image=np.meshgrid(X, Y) #
c=real+image*1j #
'''
frompyfunc(func, nin, nout)
funcninnout,
frompyfuncPythonufunc
'''
mandelbrot_set = np.frompyfunc(iterator, 3, 1)(c, 1.5, 100).astype(np.float) 
end_time = time.time()
secs = end_time - start_time
print("It took", secs, "seconds")

plt.figure(dpi=600) #dpi
plt.imshow(mandelbrot_set,extent=[-1.35, 1.35, -1.25, 1.25]) #extent
#plt.axis('off') #
plt.show()
```

```python
import time

import matplotlib.pyplot as plt #
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D #3d
import numpy as np #

start_time = time.time()

#start generating points
x_lim=np.linspace(-10,10,520)
y_lim=np.linspace(-10,10,520)
z_lim=np.linspace(-10,10,520)
X_points=[] #X
Y_points=[] #Y
Z_points=[] #Z
for x in x_lim:
    for y in y_lim:
        for z in z_lim:
            if (x**2+(9/4)*y**2+z**2-1)**3-(9/80)*y**2*z**3-x**2*z**3<=0:
                X_points.append(x)
                Y_points.append(y)
                Z_points.append(z)

end_time = time.time()
secs = end_time - start_time
print("It took", secs, "seconds")

###start plot love 
fig=plt.figure() #
ax=fig.add_subplot(111,projection='3d') #3d
ax.scatter(X_points,Y_points,Z_points,s=20,alpha=0.5,color="red") #3d
plt.show()
```
